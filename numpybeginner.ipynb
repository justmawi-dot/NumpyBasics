```python
arr = np.array([7, 4, 1, 12, 9, 0])
for x in arr:
    if x % 2 == 0:
        print(x)
```

```python
arr = np.array([7, 4, 1, 12, 9, 0])

# Use np.where to replace odd numbers with -1
arr = np.where(arr % 2 == 1, -1, arr)

print(arr)
```

```python
arr = np.array([7, 4, 1, 12, 9, 0])

reshaped_arr_2_3 = arr.reshape(2, 3)
print(f"Reshaped to (2, 3):\n{reshaped_arr_2_3}")

reshaped_arr_3_2 = arr.reshape(3, 2)
print(f"\nReshaped to (3, 2):\n{reshaped_arr_3_2}")

# Flattening the array returns a 1D array
flattened_arr = arr.flatten()
print(f"\nFlattened array: {flattened_arr}")
```
