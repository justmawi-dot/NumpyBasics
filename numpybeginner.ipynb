```python
arr = np.array([7, 4, 1, 12, 9, 0])
for x in arr:
    if x > 5:
        print(x)
```

```python
arr = np.array([7, 4, 1, 12, 9, 0])
for x in arr:
    if x % 2 == 0:
        print(x)
```

```python
arr = np.array([7, 4, 1, 12, 9, 0])

# Use np.where to replace odd numbers with -1
arr = np.where(arr % 2 == 1, -1, arr)

print(arr)
```

```python
arr = np.array([7, 4, 1, 12, 9, 0])

reshaped_arr_2_3 = arr.reshape(2, 3)
print(f"Reshaped to (2, 3):\n{reshaped_arr_2_3}")

reshaped_arr_3_2 = arr.reshape(3, 2)
print(f"\nReshaped to (3, 2):\n{reshaped_arr_3_2}")

# Flattening the array returns a 1D array
flattened_arr = arr.flatten()
print(f"\nFlattened array: {flattened_arr}")
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0,10, num=300)
y = np.sin(x)

plt.figure()
plt.plot(x,y)
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("y")
```

```python
fig,ax = plt.subplots()
ax.plot(x,y)
ax.set_title("Sine Wave")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

```python
plt.figure()
plt.scatter(x,y)
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("y")
```

```python
cats = np.array([24, 16, 22, 18, 20])
food = np.array([12, 16, 8, 9, 19])
plt.figure()
plt.bar(cats, food)
plt.title("Cats vs Food")
plt.xlabel("cats")
plt.ylabel("food")
```
